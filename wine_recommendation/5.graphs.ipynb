```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('input_variety0314.csv')
```

```python
df.columns
```

당도

```python
x = [1,2,3,4,5]
y = np.bincount(df['sweet'])[1:]

plt.bar(x,y, align='center')
plt.xticks([1,2,3,4,5])
plt.xlim(0.4, 5.6)
plt.title('Sweetness')
plt.show()
```

바디감

```python
#바디감
x = [1,2,3,4,5]
y = np.bincount(df['body'])[1:]

plt.bar(x,y, align='center')
plt.xticks([1,2,3,4,5])
plt.xlim(0.4, 5.6)
plt.title('Body')
plt.show()
```

테이스팅 노트

```python
#테이스팅 노트
x =df.columns[5:22]
y = df[x].sum(axis=0).values

plt.barh(np.arange(1,18), y, align = 'center')
plt.yticks(np.arange(1,18), np.array(x))
plt.title('Tasting note')
plt.show()
```

포도품종

```python
x = df.columns[22:-1]
y = df[x].sum(axis=0).values

plt.barh(np.arange(1,21), y, align = 'center')
plt.yticks(np.arange(1,21), np.array(x))
plt.ylim(0,21)
plt.title('Varieties')
plt.show()
```
